```python
#libraries
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException
from selenium.common.exceptions import TimeoutException
```

```python
# Configurar o driver do navegador
driver = webdriver.Chrome()

# Navegar até a página da web
driver.get("https://www.tradeinn.com/en")

# Aguarde até que o botão de aceitar cookies seja visível
elemento_botao_cookies = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.ID, 'onetrust-accept-btn-handler'))
)

# Aguarde por 2 segundos (ajuste conforme necessário)
time.sleep(2)

# Clicar no botão de aceitar cookies
elemento_botao_cookies.click()

# Aguarde por 2 segundos (ajuste conforme necessário)
time.sleep(5)

# Encontrar o elemento <a> com a classe "contenedor_tiendas_hpt"
elemento_tienda = driver.find_element(By.XPATH, '//a[@class="contenedor_tiendas_hpt" and @href="/bikeinn/en"]')

# Clicar no elemento
elemento_tienda.click()

time.sleep(2)

# Aguarde até que o elemento seja visível e interagível usando o texto do link
elemento_roupa_homem = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.LINK_TEXT, "Men´s Clothing"))
)

# Clicar no elemento
elemento_roupa_homem.click()

time.sleep(2)

# Aguardar até que o elemento "View all products" seja visível
elemento_todos_os_produtos = WebDriverWait(driver, 10).until(
    EC.visibility_of_element_located((By.LINK_TEXT, 'View all products'))
)

# Aguardar até que a sobreposição desapareça
WebDriverWait(driver, 30).until_not(
    EC.visibility_of_element_located((By.CLASS_NAME, 'barra_black_friday'))
)

# Clicar no elemento "View all products"
elemento_todos_os_produtos.click()

# Aguarde alguns segundos (ajuste conforme necessário)
time.sleep(5)

# Placeholder for data collection loop


dados_coletados = []

# Função para rolar até o final da página
def scroll_down():
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    WebDriverWait(driver, 10).until(EC.invisibility_of_element_located((By.XPATH, '//div[@class="loading"]')))

# Coletar dados enquanto houver mais conteúdo
while True:
    try:
        # Collect data for Elemento 1
        elemento_1_imagem = driver.find_element(By.XPATH, '/html/body/div[10]/div[1]/div[3]/div[5]/div[2]/div/ul/li[8]/div/div[1]/a/img').get_attribute('src')
        
        # Collect data for Elemento 2
        elemento_2 = driver.find_element(By.XPATH, '/html/body/div[10]/div[1]/div[3]/div[5]/div[2]/div/ul/li[12]/div/div[2]/h3/a').text
        
        # Collect data for Elemento 3
        elemento_3 = driver.find_element(By.XPATH, '/html/body/div[10]/div[1]/div[3]/div[5]/div[2]/div/ul/li[12]/div/div[2]/p').text
        
        # Append the collected data to the list
        dados_coletados.append([elemento_1_imagem, elemento_2, elemento_3])
        
        # Scroll down to load more content
        scroll_down()
        
    except NoSuchElementException:
        # If no more elements to collect, exit the loop
        break

# Criar um DataFrame a partir dos dados coletados
Bike_inn = pd.DataFrame(dados_coletados, columns=['Image', 'Description', 'Price'])
print(Bike_inn)

# Close the WebDriver
driver.quit()
```

```python
# Configurar o driver do navegador
driver = webdriver.Chrome()

# Navegar até a página da web
driver.get("https://www.tradeinn.com/swiminn/en")

# Aguarde até que o botão de aceitar cookies seja visível
elemento_botao_cookies = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.ID, 'onetrust-accept-btn-handler'))
)

# Aguarde por 2 segundos (ajuste conforme necessário)
time.sleep(2)

# Clicar no botão de aceitar cookies
elemento_botao_cookies.click()

# Aguarde por 2 segundos (ajuste conforme necessário)
time.sleep(5)


# Aguarde até que o elemento seja visível e interagível usando o texto do link
Women_swimwear = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.LINK_TEXT, "Women´s Swimwear"))
)
time.sleep(2)

# Clicar no elemento
Women_swimwear.click()

time.sleep(2)

# Aguardar até que o elemento "View all products" seja visível
elemento_todos_os_produtos = WebDriverWait(driver, 10).until(
    EC.visibility_of_element_located((By.LINK_TEXT, 'View all products'))
)
time.sleep(2)

# Clicar no elemento "View all products"
elemento_todos_os_produtos.click()

# Aguarde alguns segundos (ajuste conforme necessário)
time.sleep(5)

# Placeholder for data collection loop


dados_coletados = []

# Função para coletar dados de um elemento
def collect_data(element):
    try:
        elemento_imagem = element.find_element(By.XPATH, './div/div[1]/a/img').get_attribute('src')
        elemento_2 = element.find_element(By.XPATH, './div/div[2]/h3/a').text
        elemento_3 = element.find_element(By.XPATH, './div/div[2]/p').text
        dados_coletados.append([elemento_imagem, elemento_2, elemento_3])
    except NoSuchElementException:
        print("Elemento não encontrado")

# Coletar dados enquanto houver mais conteúdo
for _ in range(5):  # Defina um número fixo de rolagens, ajuste conforme necessário
    try:
        # Localizar todos os elementos na página
        elementos = driver.find_elements(By.XPATH, '/html/body/div[10]/div[1]/div[3]/div[5]/div[2]/div/ul/li')

        # Coletar dados para cada elemento
        for elemento in elementos:
            collect_data(elemento)

        # Realizar o scroll para baixo
        scroll_down()

    except NoSuchElementException:
        # Se não houver mais elementos para coletar, sair do loop
        break
    except Exception as e:
        print(f"Erro: {e}")

# Criar um DataFrame a partir dos dados coletados
Swim_inn_Women_clothes = pd.DataFrame(dados_coletados, columns=['Image', 'Description', 'Price'])
print(Swim_inn_Women_clothes)


# Close the WebDriver
driver.quit()
```

```python
Swim_inn_Women_clothes.to_csv('Swim_inn_Women_clothes.csv', index=False)
```
