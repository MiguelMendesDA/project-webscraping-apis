# books

```python
books =pd.read_csv('book_items.csv')
```

```python
books['group'] = 'books'
new_order = ['group','category', 'author' ,'Title', 'Image', 'Price']
books = books[new_order]
books.rename(columns={'Title': 'description/title', 'Image': 'image', 'Price' : 'suggested_price(€)'}, inplace=True)
```

```python
books['suggested_price(€)'] = pd.to_numeric(books['suggested_price(€)'], errors='coerce')

books = books.dropna()

books = books.applymap(lambda x: x.lower() if isinstance(x, str) else x)

books= books.reset_index(drop=True)
```
